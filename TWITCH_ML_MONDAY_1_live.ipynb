```
#colab.research.google.com

import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt

x = np.linspace(-10,10, 100)
y = np.sin(2*x)


plt.figure(figsize=[10,10])
plt.plot(x, y, color="red", marker='o')

x_real = np.array([x +(1 if random.random() < 0.5 else -1) * random.random() *2 for x in x])
y_real = np.array([y +(1 if random.random() < 0.5 else -1) * random.random() *2 for y in y])

plt.plot(x_real, y_real, marker='x')
```

```
from sklearn.metrics import mean_squared_error


def f(m,x,b):
  return m * x + b

m = 0
b = 0

y_pred = [f(m,x,b) for x in x_real]
plt.figure(figsize=[10,10])
plt.plot(x_real, y_real, color = 'blue',marker='x')
plt.plot(x_real, y_pred, color='red',marker = 'o')

# error = sum((y_pred - y_real)^2) / total_

print("mean squared error: ", mean_squared_error(y_real, y_pred))
```

```
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(X=x_real.reshape(-1,1), y=y_real.reshape(-1,1))
y_pred = reg.predict(x_real.reshape(-1,1))

plt.figure(figsize=[10,10])
plt.plot(x_real, y_real, color = 'blue',marker='x')
plt.plot(x_real, y_pred, color='red',marker = 'o')
```

```
print("Coef: ",reg.coef_)
print("Intercept: ", reg.intercept_)
print("mean_squared_error: ", mean_squared_error(y_real, y_pred))
```

```
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
errors = []
for degrees in range(1,102):
  model = make_pipeline(PolynomialFeatures(degrees), linear_model.LinearRegression())
  model.fit(x_real.reshape(-1,1), y_real.reshape(-1,1))
  y_plot = model.predict(x_real.reshape(-1,1))
  # plt.figure(figsize=[10,10])
  # plt.plot(x_real, y_real, color = 'blue',marker='x')
  # plt.plot(x_real, y_plot, color='green', marker='o')
  # plt.show()

  # print("Degree: ", degrees)
  # print("Error: ", mean_squared_error(y_real, y_plot))
  errors.append(mean_squared_error(y_real, y_plot))
  print()
  print()
```

```
plt.figure(figsize=[10,10])
plt.plot(range(1,102), errors, color = 'blue',marker='x')
plt.show()
```
